```python
import osmnx as ox

def fetch_osm_graph_from_place(place_name):
    # Télécharger le graphe de la zone spécifiée
    graph = ox.graph_from_place(place_name, network_type='drive')
    return graph

osm_graph = fetch_osm_graph_from_place("Fès, Morocco")
```

```python
def extract_nodes_and_edges(graph):
    nodes, edges = ox.graph_to_gdfs(graph)
    nodes = nodes.reset_index()
    edges = edges.reset_index()
    return nodes, edges

nodes, edges = extract_nodes_and_edges(osm_graph)
```

```python
from neo4j import GraphDatabase

class Neo4jHandler:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_node(self, tx, node_id, lat, lon):
        tx.run("CREATE (n:Node {id: $id, lat: $lat, lon: $lon})", id=node_id, lat=lat, lon=lon)

    def create_relationship(self, tx, start_node_id, end_node_id):
        tx.run("""
        MATCH (a:Node {id: $start_id}), (b:Node {id: $end_id})
        CREATE (a)-[:CONNECTED_TO]->(b)
        """, start_id=start_node_id, end_id=end_node_id)

def import_to_neo4j(nodes, edges, neo4j_uri, user, password):
    neo4j_handler = Neo4jHandler(neo4j_uri, user, password)

    with neo4j_handler.driver.session() as session:
        for _, row in nodes.iterrows():
            session.write_transaction(neo4j_handler.create_node, row['osmid'], row['y'], row['x'])

        for _, row in edges.iterrows():
            session.write_transaction(neo4j_handler.create_relationship, row['u'], row['v'])

    neo4j_handler.close()
neo4j_uri = "bolt://localhost:7687"
user = "neo4j"
password = "0606383390"
import_to_neo4j(nodes, edges, neo4j_uri, user, password)
```

```python
import osmnx as ox

def fetch_hotels_from_place(place_name):
    tags = {'tourism': 'hotel'}
    hotels = ox.geometries_from_place(place_name, tags)
    return hotels

hotels = fetch_hotels_from_place("Fès, Morocco")
```

```python
def import_hotels_to_neo4j(hotels, neo4j_uri, user, password):
    neo4j_handler = Neo4jHandler(neo4j_uri, user, password)

    with neo4j_handler.driver.session() as session:
        for _, hotel in hotels.iterrows():
            node_id = hotel['name']  # Utilisation de la colonne 'name' comme identifiant unique
            # Calculer les coordonnées du centre du polygone
            lat, lon = hotel['geometry'].centroid.y, hotel['geometry'].centroid.x
            session.write_transaction(neo4j_handler.create_node, node_id, lat, lon)

    neo4j_handler.close()

import_hotels_to_neo4j(hotels, neo4j_uri, user, password)
```

```python
import osmnx as ox
from geopy.distance import great_circle
from neo4j import GraphDatabase
import pandas as pd
# Fonction pour calculer la distance entre deux points géographiques en mètres
def calculate_distance(lat1, lon1, lat2, lon2):
    point1 = (lat1, lon1)
    point2 = (lat2, lon2)
    return great_circle(point1, point2).meters

# Fonction pour créer des relations entre les hôtels et les nœuds routiers les plus proches
def create_relationships_with_nearest_hotel(nodes, hotels):
    for _, hotel in hotels.iterrows():
        hotel_lat = hotel['geometry'].y
        hotel_lon = hotel['geometry'].x

        # Trouver l'hôtel le plus proche
        nearest_node = None
        min_distance = float('inf')
        for _, node in nodes.iterrows():
            node_lat = node['y']
            node_lon = node['x']
            distance = calculate_distance(hotel_lat, hotel_lon, node_lat, node_lon)
            if distance < min_distance:
                min_distance = distance
                nearest_node = node['osmid']

        # Créer une relation entre l'hôtel et le nœud routier le plus proche
        create_relationship_with_nearest_hotel(hotel['osmid'], nearest_node)

# Fonction pour ajouter des informations supplémentaires sur les hôtels dans Neo4j
def add_hotel_information_to_neo4j(hotels, neo4j_uri, user, password):
    neo4j_handler = Neo4jHandler(neo4j_uri, user, password)

    with neo4j_handler.driver.session() as session:
        for _, hotel in hotels.iterrows():
            node_id = hotel['osmid']
            hotel_name = hotel['name']
            # Ajouter d'autres informations sur l'hôtel comme propriétés du nœud
            session.write_transaction(neo4j_handler.add_hotel_information, node_id, hotel_name)

    neo4j_handler.close()
# Fonction pour extraire les nœuds routiers de Neo4j
def extract_nodes_from_neo4j(neo4j_uri, user, password):
    neo4j_handler = Neo4jHandler(neo4j_uri, user, password)
    nodes = []

    with neo4j_handler.driver.session() as session:
        result = session.run("MATCH (n:Node) RETURN n")
        for record in result:
            node = record["n"]
            nodes.append({
                'osmid': node['id'],
                'lat': node['lat'],
                'lon': node['lon']
            })

    neo4j_handler.close()
    return pd.DataFrame(nodes)


# Exemple d'utilisation
place_name = "Fès, Morocco"

# Extraction des hôtels de Fès
hotels = fetch_hotels_from_place(place_name)

# Importation des hôtels dans Neo4j
import_hotels_to_neo4j(hotels, neo4j_uri, user, password)

# Extraction des nœuds routiers déjà présents dans Neo4j (à partir de l'étape précédente)
nodes = extract_nodes_from_neo4j(neo4j_uri, user, password)

# Création de relations entre les hôtels et les nœuds routiers les plus proches
create_relationships_with_nearest_hotel(nodes, hotels)

# Ajout d'informations supplémentaires sur les hôtels dans Neo4j
add_hotel_information_to_neo4j(hotels, neo4j_uri, user, password)
```
