```python
import osmnx as ox
import pandas as pd
from shapely.geometry import Point

def fetch_osm_graph_from_place(place_name):
    # Tlcharger le graphe de la zone spcifie
    graph = ox.graph_from_place(place_name, network_type='drive')
    return graph

def fetch_hotels_from_place(place_name):
    # Tlcharger les POI pour les htels
    tags = {'tourism': 'hotel'}
    hotels = ox.geometries_from_place(place_name, tags)
    return hotels

def add_hotels_to_graph(graph, hotels):
    # Ajouter les htels au graphe avec leurs noms et nuds
    for idx, hotel in hotels.iterrows():
        if isinstance(hotel.geometry, Point):  # Assurez-vous de traiter uniquement les Points
            if 'name' in hotel:
                node_id = ox.distance.nearest_nodes(graph, hotel.geometry.x, hotel.geometry.y)
                graph.nodes[node_id]['hotel_name'] = hotel['name']
                graph.nodes[node_id]['hotel'] = True
            else:
                node_id = ox.distance.nearest_nodes(graph, hotel.geometry.x, hotel.geometry.y)
                graph.nodes[node_id]['hotel'] = True
    return graph

place_name = "Fs, Morocco"
osm_graph = fetch_osm_graph_from_place(place_name)
hotels = fetch_hotels_from_place(place_name)
osm_graph = add_hotels_to_graph(osm_graph, hotels)
```

```python
def graph_to_dataframe(graph):
    # Convertir le graphe en DataFrame pour les nuds et les artes
    nodes = pd.DataFrame.from_dict(dict(graph.nodes(data=True)), orient='index')
    edges = ox.graph_to_gdfs(graph, nodes=False).reset_index()[['u', 'v']]
    return nodes.reset_index().rename(columns={'index': 'osmid'}), edges

nodes, edges = graph_to_dataframe(osm_graph)
```

```python
from neo4j import GraphDatabase

class Neo4jHandler:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_node(self, tx, node_id, lat, lon, hotel_name=None, hotel=False):
        query = "CREATE (n:Node {id: $id, lat: $lat, lon: $lon, hotel: $hotel"
        if hotel_name:
            query += ", hotel_name: $hotel_name"
        query += "})"
        tx.run(query, id=node_id, lat=lat, lon=lon, hotel=hotel, hotel_name=hotel_name)

    def create_relationship(self, tx, start_node_id, end_node_id):
        tx.run("""
        MATCH (a:Node {id: $start_id}), (b:Node {id: $end_id})
        CREATE (a)-[:CONNECTED_TO]->(b)
        """, start_id=start_node_id, end_id=end_node_id)

def import_to_neo4j(nodes, edges, neo4j_uri, user, password):
    neo4j_handler = Neo4jHandler(neo4j_uri, user, password)

    with neo4j_handler.driver.session() as session:
        for _, row in nodes.iterrows():
            session.write_transaction(
                neo4j_handler.create_node, 
                row['osmid'], 
                row['y'], 
                row['x'], 
                row.get('hotel_name', None), 
                row.get('hotel', False)
            )

        for _, row in edges.iterrows():
            session.write_transaction(neo4j_handler.create_relationship, row['u'], row['v'])

    neo4j_handler.close()

neo4j_uri = "bolt://localhost:7687"
user = "neo4j"
password = "0608410180"
import_to_neo4j(nodes, edges, neo4j_uri, user, password)

print("Graphe et htels ajouts  Neo4j avec succs!")
```

```python
import requests
from geopy.distance import geodesic
from neo4j import GraphDatabase
import folium
import webbrowser

class Neo4jHandler:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def get_hotels(self):
        with self.driver.session() as session:
            result = session.run("MATCH (n:Node) WHERE n.hotel = true RETURN n.id as id, n.lat as lat, n.lon as lon, n.hotel_name as name")
            return [{"id": record["id"], "lat": record["lat"], "lon": record["lon"], "name": record["name"]} for record in result]

def get_location():
    response = requests.get('http://ip-api.com/json/')
    data = response.json()
    latitude = data['lat']
    longitude = data['lon']
    return latitude, longitude

def find_nearest_hotel(hotels, lat, lon):
    nearest_hotel = None
    min_distance = float('inf')

    for hotel in hotels:
        hotel_location = (hotel['lat'], hotel['lon'])
        distance = geodesic((lat, lon), hotel_location).meters
        if distance < min_distance:
            min_distance = distance
            nearest_hotel = hotel

    return nearest_hotel, min_distance

def get_route(start, end):
    url = f'http://router.project-osrm.org/route/v1/driving/{start[1]},{start[0]};{end[1]},{end[0]}?overview=full&geometries=geojson'
    response = requests.get(url)
    data = response.json()
    if 'routes' in data and data['routes']:
        route = data['routes'][0]['geometry']['coordinates']
        return [(coord[1], coord[0]) for coord in route]
    else:
        raise Exception("No route found")

# Configuration Neo4j
neo4j_uri = "bolt://localhost:7687"
user = "neo4j"
password = "0606383390"

# Initialisation de la connexion Neo4j
neo4j_handler = Neo4jHandler(neo4j_uri, user, password)

# Rcupration des htels depuis Neo4j
hotels = neo4j_handler.get_hotels()

# Fermeture de la connexion Neo4j
neo4j_handler.close()

# Obtention de la position actuelle
latitude, longitude =(34.03715133666992,4.999800205230713)
print(f"Votre position actuelle: Latitude: {latitude}, Longitude: {longitude}")

# Recherche de l'htel le plus proche
nearest_hotel, distance = find_nearest_hotel(hotels, latitude, longitude)
if nearest_hotel:
    print(f"L'htel le plus proche est '{nearest_hotel['name']}'  une distance de {distance:.2f} mtres.")
    
    # Cration de la carte
    map_ = folium.Map(location=[latitude, longitude], zoom_start=15)

    # Ajout d'un marqueur pour l'emplacement actuel
    folium.Marker(
        location=[latitude, longitude],
        popup='Votre position actuelle',
        icon=folium.Icon(color='blue')
    ).add_to(map_)

    # Ajout d'un marqueur pour l'htel le plus proche
    folium.Marker(
        location=[nearest_hotel['lat'], nearest_hotel['lon']],
        popup=f"Htel: {nearest_hotel['name']}",
        icon=folium.Icon(color='red')
    ).add_to(map_)

    # Obtention de l'itinraire entre l'emplacement actuel et l'htel le plus proche
    try:
        route = get_route((latitude, longitude), (nearest_hotel['lat'], nearest_hotel['lon']))

        # Ajout du chemin entre l'emplacement actuel et l'htel le plus proche
        folium.PolyLine(
            locations=route,
            color='green'
        ).add_to(map_)
    except Exception as e:
        print(f"Erreur lors de la rcupration de l'itinraire: {e}")

    # Sauvegarde de la carte dans un fichier HTML
    map_file = 'map.html'
    map_.save(map_file)
    print(f"Carte sauvegarde sous le nom '{map_file}'. Ouvrez ce fichier dans un navigateur pour voir la carte.")

    # Ouverture automatique de la carte dans le navigateur
    webbrowser.open(map_file)
else:
    print("Aucun htel trouv  proximit.")
```
